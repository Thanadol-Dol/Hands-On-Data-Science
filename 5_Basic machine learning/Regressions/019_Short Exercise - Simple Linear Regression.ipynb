# Short Exercise: Simple Linear Regression

In this exercise, you'll build a very simple linear regression model that predicts salary of a person given his/her years of experience.

The given file, Salary_Data.xlsx, only have 2 columns
- YearsExperience :Years of experience of a person 
- Salary :Salary of a person

Feel free to use more/less cells than provided

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

Import Salary_Data.xlsx here

```python
df = pd.read_excel("Salary_Data.xlsx")
```

```python
df.head()
```

```python
X = df[['YearsExperience']]
y = df['Salary']
```

## Splitting the dataset into the Training set and Test set

Use 75% training 25% testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

## Predicting the Test set results

```python
y_pred_train = model.predict(X_train)
```

```python
y_pred = model.predict(X_test)
```

## Visualising the Training set results

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, y_pred_train, color = 'blue')
plt.title('Salary vs YearsExperience (Train)')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

## Visualising the Test set results

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_test, y_pred, color = 'blue')
plt.title('Salary vs YearsExperience (Test)')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.show()
```

## Evaluating how the model performs

```python
from sklearn.metrics import r2_score

r2_score(y_test, y_pred) 
```

How does the model perform this time around?

```python
r2_score(y_train,y_pred_train)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
