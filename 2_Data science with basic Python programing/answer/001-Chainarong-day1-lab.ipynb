# Lab6: Grading

```python
def show_student(name,score,organization="No organization"):
  print("Name:",name)
  print("Organization:",organization)
  print("Score:",score)
  percent = int(score)*100/88
  print("Score (%): {:.1f}".format(percent))
  grade = ''
  if percent>80:
    grade = 'A'
  elif percent>=60 and percent<=80: 
    grade = 'B'
  elif percent>=40 and percent<60: 
    grade = 'C'
  else:
    grade = 'D'
    
  print("Grade:",grade)
  print('\n') #Enter

score = input('Enter score: ')

show_student(organization="GBDi", name="Mr.A",score=score)
```

# LAB8: String Combos

```python
x = 'Python is fun!'
x[::-1]
```

```python
def remove_title(name):
  name = name.strip()  
  if(name[0:6]=='นางสาว'):
    name = name[6:]
  elif(name[0:4]=='Mrs.' ):
    name = name[4:] 
  elif(name[0:3]=='Mr.' or name[0:3]=='นาย' or name[0:3]=='นาง' ):
    name = name[3:]   

  name = name.strip()
  return name

new_name = remove_title(' Mr. Pongsakorn  ')
print(new_name)
print(remove_title('นางสาว สมศรี นายนารัตน์  '))
print(remove_title('นายนายะ มาโตยูจิ '))
print(remove_title('นายะ มาโตยูจิ '))
```

# LAB10: dictionary

```python
def convert_month(date_string):
  # 19-05-2564  
  date = date_string[0:2]
  month = date_string[3:5]
  year = date_string[6:11]
  month_thai = {
      '01' : 'มกราคม',      '02' : 'กุมภาพันธ์',      '03' : 'มีนาคม',
      '04' : 'เมษายน',      '05' : 'พฤษภาคม',      '06' : 'มิถุนายน',
      '07' : 'กรกฎาคม',      '08' : 'สิงหาคม',      '09' : 'กันยายน',
      '10' : 'ตุลาคม',      '11' : 'พฤศจิกายน',      '12' : 'ธันวาคม',
  }

  print(int(date),month_thai[month],year)


convert_month('19-05-2564')
convert_month('09-09-2549')
convert_month('31-12-2500')
```

# LAB11: For-Loop

```python
sum = 0
for i in range(1,501):
  sum=sum+i
  # รอบที่ 1 sum = 0 + 1
  # รอบที่ 2 sum = 1 + 2
  # รอบที่ 3 sum = 3 + 3
print(sum)
```

```python
sum = 0
for i in range(0,501,2):
  sum+=i


# OR


sum = 0
for i in range(1,501,1):
  if(i%2==0):
    sum+=i
print(sum)
```

```python
def mul_table(number):
  for i in range(1,13):
    print(number,'x',i,'=',number*i)

mul_table(2)
```

```python
def a(x,y,z):
  return {"x":x+1,"y":y+1,"z":z+1}

a(1,10,100)
```

```python
def mul_tables(numbers):
  for i in numbers:
    mul_table(i)
    print('\n')

mul_tables([1,3,5])
```

# LAB12: While-Loop

```python
sum = 0
text = input("Enter positive integer: ")
while(text.isnumeric() and text!='0'):
  sum = sum + int(text)
  # 0 + int("15")
  # 15 + int("2")
  text = input("Enter positive integer: ")
print(sum)
```
