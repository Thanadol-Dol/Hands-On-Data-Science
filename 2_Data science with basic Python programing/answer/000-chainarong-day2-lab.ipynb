# LAB13 num2word

```python
def num2word(number):
  
  is_integer = str(type(number))=="<class 'int'>" #OR
  is_integer = type(number)==int
  
  # Check input condition
  if not(is_integer):
    print("กรุณาใส่จำนวนเต็มที่มีค่าไม่เกิน 1 ล้าน")
  elif number > 1000000 :
    print("กรุณาใส่จำนวนเต็มที่มีค่าไม่เกิน 1 ล้าน")
  else:

    # Zero padding
    str_number = str(number)  
    while(len(str_number)<7):
      str_number = '0'+str_number

    # Setup dictionary 
    number_thai = {
        '1':'หนึ่ง',
        '2':'สอง',
        '3':'สาม',
        '4':'สี่',
        '5':'ห้า',
        '6':'หก',
        '7':'เจ็ด',
        '8':'แปด',
        '9':'เก้า',
    }
    unit_thai = {
         0 : 'ล้าน',
         1 : 'แสน',
         2 : 'หมื่น',
         3 : 'พัน',
         4 : 'ร้อย',
         5 : 'สิบ',
         6 : ''        
    }
    text_thai = ''
    
    # Loop print
    for i in range(0,7):
      if(str_number[i]!='0'):
        # check ค่าและฐานที่ ร
        if(str_number[i]=='1' and i==6):
          text_thai += 'เอ็ด'
        elif(str_number[i]=='2' and i==5):
          text_thai += 'ยี่สิบ'
        else:
          text_thai += number_thai[str_number[i]]+unit_thai[i]

    print(text_thai)

num2word(123451)
num2word(2021)
num2word(1000001)
num2word('Hello')
```
