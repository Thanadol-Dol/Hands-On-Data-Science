# LAB13 num2word

```python
def num2word(num):
  value_list = []
  if (str(num).isnumeric() == False or num > 1000000):
    print('กรุณาใส่จำนวนเต็มที่มีค่าไม่เกิน 1 ล้าน')
  else:
    digits = {0:'', 1:'สิบ', 2:'ร้อย', 3:'พัน', 4:'หมื่น', 5:'แสน', 6:'ล้าน'}
    values = {0:'', 1:'หนึ่ง', 2:'สอง', 3:'สาม', 4:'สี่', 5:'ห้า', 6:'หก', 7:'เจ็ด', 8:'แปด', 9:'เก้า'}
    while(int(num) > 0):
      value_list.append(int(num%10))
      num = num/10
    for i in reversed(range(len(value_list))):
      if (value_list[i] == 0) :
        continue
      if (i == 1 and value_list[i] == 2) :
        print("ยี่" + digits[i],end='')
        continue
      if (i == 0 and value_list[i] == 1 and len(value_list) > 1) :
        print("เอ็ด",end='')
        break
      else :
        print(values[value_list[i]] + digits[i],end='')
    return value_list
num2word(123451)
print('\n')
num2word(2021)
print('\n')
num2word(1000001)
print('\n')
num2word('Hello')
```
