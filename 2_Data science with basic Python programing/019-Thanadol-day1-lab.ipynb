# Lab6: Grading

```python
def show_student(name, organization="No organization"):
  score_input = int(input("Enter score: "))
  percent = round((score_input/88)*100,1)
  print("Name:",name)
  print("Organization:",organization)
  print("Score:",score_input)
  print("Score (%):",percent)
  if percent > 80:
    print("Grade: A")
  elif percent >= 60 and percent <= 80:
    print("Grade: B")
  elif percent >= 40 and percent < 60:
    print("Grade: C")
  else:
    print("Grade: D")
  print('\n')
show_student(organization="GBDi", name="Mr.A")
```

# LAB10: dictionary

```python
def convert_month(date):
  month = {'01' : 'มกราคม', '02' : 'กุมภาพันธ์', '03' : 'มีนาคม', '04' : 'เมษายน', '05' : 'พฤษภาคม', '06' : 'มิถุนายน', '07' : 'กรกฎาคม', '08' : 'สิงหาคม', '09' : 'กันยายน', '10' : 'ตุลาคม', '11' : 'พฤศจิกายน', '12' : 'ธันวาคม'}
  temp = date.split('-')
  print(int(temp[0]),month[temp[1]],temp[2])
```

```python
convert_month('19-05-2564')
convert_month('09-09-2549')
convert_month('31-12-2500')
```

```python
def mul_table(num):
  for i in range(1,13):
    print('{0} x {1} = {2}'.format(num,i,num*i))
  print('\n')
mul_table(2)
mul_table(25)
```

```python
def mul_tables(num_list):
  for i in num_list:
    for j in range(1,13):
      print('{0} x {1} = {2}'.format(i,j,i*j))
    print('\n')
mul_tables([1,3,5])
```
